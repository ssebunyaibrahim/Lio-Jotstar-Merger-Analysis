## **1. Total Users & Growth Trends**

```python
# Total Users
users_lio= ls.user_id.nunique()
users_jot = js.user_id.nunique()
users_diff= users_lio-users_jot
users_per_diff =round((users_diff/users_lio)*100,2)

print("Total Users Liocinema: ",users_lio)
print("Total Users jotstar: ",users_jot)
print("Difference: ",users_diff)
print(f"Percentage_Difference: {users_per_diff}%")
```

ðŸ“Œ LioCinema is pulling ahead as the dominant OTT platform among the two, having acquired 138,826(75.68%) more users than Jotstar over the year.  

ðŸ“Œ Jotstarâ€™s total user base remains significantly lower at 4,4620, reflecting limited growth potential or a lack of aggressive acquisition efforts.

```python
# Extract year-month from subscription date
ls["month"]= ls.subscription_date.dt.to_period("M").astype(str)
js["month"]= js.subscription_date.dt.to_period("M").astype(str)

# Count new users per month
lio_growth=ls.groupby("month")["user_id"].nunique().reset_index().rename(columns={"user_id":"new_users_lio"})
jot_growth=js.groupby("month")["user_id"].nunique().reset_index().rename(columns={"user_id":"new_users_jot"})

# Merge both data frames
growth_trends= pd.merge(lio_growth,jot_growth,on="month")
```

```python
# Create Mapping for month names
months={"2024-01":'January',"2024-02":'February','2024-03':'March',"2024-04":'April',"2024-05":'May',"2024-06":'June',
        "2024-07":'July',"2024-08":'August',"2024-09":'September',"2024-10":'October',"2024-11":'November',"2024-12":'December'}
growth_trends.month=growth_trends['month'].map(months)
print('User Growth Trends')
print(growth_trends)
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=growth_trends,x="month",y="new_users_lio",label="lioCinema",color='#023047',marker='o')
sns.lineplot(data=growth_trends,x="month",y="new_users_jot",label="Jotstar",marker='s',linestyle='--')
plt.xticks(rotation=45)
plt.title("User Growth Trends (January - November 2024)")
plt.xlabel("Month (2024)")
plt.ylabel("New Users")
plt.savefig("User Growth Trends (January - November 2024).png",dpi=300,bbox_inches='tight')
plt.show()
```

ðŸ“Œ LioCinema has demonstrated exponential growth, starting with 6,758 new users in January 2024 and surging to 36,997 by November 2024, reflecting an almost 5.5x increase over the 11-month period.  

ðŸ“Œ In contrast, Jotstar has remained relatively stable, with user acquisition fluctuating between 3,934 in January and 4,262 in November, translating to a modest 8% growth over the same period.

## **4. Active vs. Inactive Users** 

```python
# Create a new column to classify users as Active or Inactive
ls['status']=ls.last_active_date.isnull().map({True: "Active", False: "Inactive"})
js['status']=js.last_active_date.isnull().map({True: "Active", False: "Inactive"})

# Calculate the percentage of Active vs. Inactive users
active_ls=round(ls.status.value_counts(normalize=True)*100,2)
active_js=round(js.status.value_counts(normalize=True)*100,2)

# Combine both series
status=pd.DataFrame(
    {'Liocinema':active_ls,
     'Jotstar':active_js}
)

print("Active Vs Inactive Users Overall (%)")
print(status)
```

```python
# Group by Age Group and 
lio_age_activity=pd.crosstab(ls['age_group'],ls['status'],normalize='index')*100
jot_age_activity=pd.crosstab(js['age_group'],js['status'],normalize='index')*100

# Display results
print("\nLioCinema - Active vs. Inactive Users by Age Group (%)")
print(lio_age_activity)

print("\nJotstar - Active vs. Inactive Users by Age Group (%)")
print(jot_age_activity)
```

```python
# Group by Subscription Plan and Status
lio_plan_activity = pd.crosstab(ls['subscription_plan'],ls['status'],normalize='index')*100
jotstar_plan_activity = pd.crosstab(js['subscription_plan'],js['status'],normalize=True)*100

# Display results
print("\nLioCinema - Active vs. Inactive Users by Subscription Plan (%)")
print(lio_plan_activity)

print("\nJotstar - Active vs. Inactive Users by Subscription Plan (%)")
print(jotstar_plan_activity)
```

```python
# Plot the Active vs. Inactive user distribution
fig=plt.figure(figsize=(24,12))
gs=fig.add_gridspec(nrows=3,ncols=2)
ax1=fig.add_subplot(gs[0,:])
ax2=fig.add_subplot(gs[1,0])
ax3=fig.add_subplot(gs[1,1:])
ax4=fig.add_subplot(gs[2,0])
ax5=fig.add_subplot(gs[2,1])

# Plot bar chart for Overall status
status.plot(kind='bar', color=['#023047','#00674F'], figsize=(12, 12),ax=ax1)
for container in ax1.containers:
    ax1.bar_label(container,fmt='%.2f%%',fontsize=15)
ax1.set_title("Active vs. Inactive Users (%)")
ax1.set_xlabel("Status")
ax1.set_ylabel("Percentage")
ax1.tick_params(axis='x',rotation=0)
ax1.yaxis.set_visible(False)

# Plot Active vs. Inactive users by Age Group for each platform
sns.heatmap(lio_age_activity,ax=ax2,cmap='Blues',linecolor='white',linewidths=0.5,annot=True,fmt='.1f')
ax2.set_title("LioCinema")

sns.heatmap(jot_age_activity,ax=ax3,annot=True,fmt='.1f',linecolor='white',linewidths=0.5, cmap='Blues',)
ax3.set_title("Jotstar")


# Plot Active vs. Inactive users by Subscription Plan for each platform
sns.heatmap(lio_plan_activity, annot=True, cmap='Purples', fmt=".1f", ax=ax4,linecolor='white',linewidths=0.5)
ax4.set_title("LioCinema")

sns.heatmap(jotstar_plan_activity, annot=True, cmap='Purples', fmt=".1f", ax=ax5,linecolor='white',linewidths=0.5)
ax5.set_title("Jotstar")

plt.tight_layout()
plt.savefig("Active vs Inactive Users.png", dpi=300, bbox_inches='tight')
plt.show()
```

## **5. Watch Time Analysis** 

```python
#Convert Watchtime to hrs
lcc['total_watch_time_hrs']=lcc['total_watch_time_mins']/60
jcc['total_watch_time_hrs']=jcc['total_watch_time_mins']/60

# Compute overall average watch time
lio_avg_watch = lcc["total_watch_time_hrs"].mean()
jot_avg_watch = jcc["total_watch_time_hrs"].mean()

# Create DataFrame for visualization
platform_watch_time = pd.DataFrame({
    "Platform": ["LioCinema", "Jotstar"],
    "Avg Watch Time": [lio_avg_watch, jot_avg_watch]
})

print(platform_watch_time)
```

```python
#  Average watch time by city_tier
# Merge watch data with subscriber info
lio_merged= pd.merge(lcc,ls,on='user_id')
jot_merged=pd.merge(jcc,js,on='user_id')

# Compute average watch time per city tier (in hours)
lio_city_watch=lio_merged.groupby('city_tier')['total_watch_time_hrs'].mean()
jot_city_watch=jot_merged.groupby('city_tier')['total_watch_time_hrs'].mean()

#Create dataframes for comparison
city_watch=pd.DataFrame({
    'Liocinema_Avg_Watch_hrs':lio_city_watch,
    'Jotstar_Avg_Watch_hrs':jot_city_watch
})

print("Average watch time by city_tier")
print(city_watch)
```

```python
#Average watch_time by device_type

# Compute average watch time per content_type (in hours)
lio_device_watch=lio_merged.groupby('device_type')['total_watch_time_hrs'].mean()
jot_device_watch=jot_merged.groupby('device_type')['total_watch_time_hrs'].mean()

#Create dataframes for comparison
device_watch=pd.DataFrame({
    'Liocinema_Avg_Watch_hrs':lio_device_watch,
    'Jotstar_Avg_Watch_hrs':jot_device_watch
})

print("Average watch time by device_type")
print(device_watch)
```

```python
# Plot watchtime analysis
fig=plt.figure(figsize=(12,6))
gs=fig.add_gridspec(nrows=3,ncols=1)
ax1=fig.add_subplot(gs[0])
ax2=fig.add_subplot(gs[1])
ax3=fig.add_subplot(gs[2])


# Plot bar chart for Overall watchtime
sns.barplot(platform_watch_time,x='Avg Watch Time',ax=ax1,hue='Platform',palette=['#023047','#219EBC'])
for container in ax1.containers:
    ax1.bar_label(container, fmt='%.2f', fontsize=15)

ax1.set_title("Overall Average Watchtime (hrs)")
ax1.set_xlabel("Avg_Hrs")
ax1.set_ylabel("Platform")
ax1.tick_params(axis='x',rotation=0)
ax1.xaxis.set_visible(False)

#Plot bar chart for city_tier watchtime
city_watch.plot(kind='bar',ax=ax2,color=['#023047','#219EBC'],figsize=(12,10))
for container in ax2.containers:
    ax2.bar_label(container,fmt='%.2f')
ax2.set_title("City_tier Watchtime (hrs)")
ax2.set_xlabel("City_tier")
ax2.set_ylabel("Avg_Hrs")
ax2.tick_params(axis='x',rotation=0)
ax2.legend(labels=["LioCinema","Jotstar"])

# Plot bar chart for device_type watchtime
device_watch.plot(kind='bar',ax=ax3,color=['#023047','#219EBC'],figsize=(12,10))
for container in ax3.containers:
    ax3.bar_label(container,fmt='%.2f')
ax3.set_title("Device_type Watchtime (hrs)")
ax3.set_xlabel("Device_type")
ax3.set_ylabel("Avg_Hrs")
ax3.tick_params(axis='x',rotation=0)
ax3.legend(labels=["LioCinema","Jotstar"])

plt.tight_layout()
plt.savefig("Watchtime Analysis.png", dpi=300, bbox_inches='tight')
plt.show()
```

## **7. Downgrade Trends**

```python
# Define the subscription hierarchy
hierarchy = {"Premium": 4,"VIP": 3, "Basic": 2, "Free": 1}

# create dataframes for Upgrade and downgrade Trends analysis
lud=ls.dropna(subset=['new_subscription_plan'])
jud=js.dropna(subset=['new_subscription_plan'])

# Add numeric rankings for easier comparison
lud['subscription_plan_rank'] = lud.subscription_plan.map(hierarchy)
lud['new_subscription_plan_rank'] = lud.new_subscription_plan.map(hierarchy)

jud['subscription_plan_rank'] = jud.subscription_plan.map(hierarchy)
jud['new_subscription_plan_rank'] = jud.new_subscription_plan.map(hierarchy)

# Identify downgrade and upgrade cases
lud['downgrade']= (lud.new_subscription_plan_rank < lud.subscription_plan_rank).astype(int)
lud['upgrade']= (lud.new_subscription_plan_rank > lud.subscription_plan_rank).astype(int)

jud['downgrade']= (jud.new_subscription_plan_rank < jud.subscription_plan_rank).astype(int)
jud['upgrade']= (jud.new_subscription_plan_rank > jud.subscription_plan_rank).astype(int)

jud.head()
```

## **10. Revenue Analysis**

```python
# Define subscription prices
lio_prices = {"Basic": 69, "Premium": 129}
jot_prices = {"VIP": 159, "Premium": 359}


# Function to generate a list of months between two dates
def get_month_range(start_date,end_date):
    months = pd.date_range(start=start_date,end=end_date,freq='MS').strftime('%Y-%m').tolist()
    return months

# Function to calculate monthly revenue breakdown
def calculate_monthly_revenue(df,prices):
    monthly_revenue={}
    
    for _,row in df.iterrows():
        start_date=row['subscription_date']
        end_date=row['last_active_date'] if pd.notna(row['last_active_date']) else pd.to_datetime(2024-11-30)
        
         # If the user changed plans
        if pd.notna(row['plan_change_date']):
            change_date=row['plan_change_date']
            
            # First plan period
            for month in get_month_range(start_date,change_date):
                monthly_revenue[month]=monthly_revenue.get(month,0) + prices.get(row['subscription_plan'],0)
            
            # Second plan period
            for month in get_month_range(change_date,end_date):
                monthly_revenue[month]=monthly_revenue.get(month,0) + prices.get(row['new_subscription_plan'],0)
        
        else:
            for month in get_month_range(start_date,end_date):
                monthly_revenue[month]= monthly_revenue.get(month,0) + prices.get(row['subscription_plan'],0)
        
    return monthly_revenue
            
 # Compute monthly revenue
lio_monthly_revenue = calculate_monthly_revenue(ls, lio_prices)
jot_monthly_revenue = calculate_monthly_revenue(js, jot_prices)     

# Convert to DataFrame for visualization
rev_lio = pd.DataFrame(list(lio_monthly_revenue.items()), columns=["Month", "Liocinema"])

rev_jot = pd.DataFrame(list(jot_monthly_revenue.items()), columns=["Month", "Jotstar"])
```

```python
# Combine dataframes for comparison
revenue=pd.merge(rev_lio,rev_jot,on= 'Month')
revenue=revenue.sort_values('Month').reset_index().drop(columns=['index'])
revenue['Month']=pd.to_datetime(revenue['Month']).dt.month_name()
print(revenue)
```

```python
# Total revenue For each Platfrom
lio_revenue=revenue.Liocinema.sum()
jot_revenue=revenue.Jotstar.sum()
total_revenue=pd.DataFrame({
    'Liocinema':lio_revenue,
    'Jotstar':jot_revenue
},index=['Total revenue'])
print(total_revenue)
```
